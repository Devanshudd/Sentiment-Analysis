```python
# Define a function to remove special characters, English characters, and consecutive newline characters
def clean_text(text):
    # Remove special characters, English characters, and consecutive newline characters
    cleaned_text = re.sub(r'[^0-9\u0A80-\u0AFF\s]', '', text)
    cleaned_text = re.sub(r'[a-zA-Z0-9]', '', cleaned_text)
    cleaned_text = re.sub(r'\n+', ' ', cleaned_text)
    return cleaned_text

# Apply the cleaning function to the 'Text' field
df['Text'] = df['Text'].apply(clean_text)
print(df)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Read the dictionary words from the text file
with open('/content/drive/MyDrive/sentiment_accuracy/ngram_dict.txt', 'r', encoding='utf-8') as file:  # Replace 'dictionary.txt' with the actual file name and path
    dictionary_words = file.read().splitlines()

# Define the stopwords in Gujarati
stop_words = ['કા', 'પર', 'છે', 'માટે', 'કરશે', 'ક', 'ખાસ', 'આવે', 'કર્યું', 'આપ્યા', 'કારણે', 'આવેલા', 'માર', 'બની', 'જાય', 'પાડી', 'તેમને', 'કારણ', 'સમજો', 'કરતાં', 'આપીને', 'થાય', 'ધરી', 'રહો', 'કરવી', 'છેવટે', 'જેવું', 'સાથે', 'ક્યારેય', 'આવી', 'આપી', 'સાથ', 'શકે', 'જાય', 'અલગ', 'બનશે', 'વિના', 'થી', 'હતા', 'જ', 'કરી', 'નું', 'કર્યુ', 'મળી', 'હતી', 'લેતા', 'લીધો', 'છતાં', 'અને', 'કે', 'કોને', 'કોઈ', 'ક્યારે', 'ક્યાં', 'ક્યાસમ', 'માંગી', 'લાગુ', 'દાવો', 'ગયો', 'મળ્યો', 'પડી', 'ક્યોંકિ', 'ક્યોંકી', 'જે', 'જેમ', 'તે', 'તેના', 'તેની', 'તેને', 'ત્યાં', 'ત્યારે', 'ત્યે', 'થઈ', 'થતા', 'દ્વારા', 'ના', 'ને', 'પર', 'પણ', 'પાછળ', 'પોતાના', 'પોતે', 'મેં', 'યા', 'યે', 'રહ્યા', 'રહ્યાં', 'રીતે', 'હવે']

# Extract sentences from the 'Text' column of the DataFrame
sentences = df1['Text'].dropna().astype(str).tolist()

# Convert the sentences to lowercase
sentences = [sentence.lower() for sentence in sentences]

# Filter the sentences that contain at least one word from the dictionary
filtered_sentences = [sentence for sentence in sentences if any(word in sentence for word in dictionary_words)]

# Define the range of n-grams you want to generate
ngram_range = (2, 2)  # Generate bigrams

# Custom tokenizer function to exclude single-letter words and stopwords
def custom_tokenizer(text):
    words = text.split()
    filtered_words = [word for word in words if len(word) > 1 and word.lower() not in stop_words]
    return filtered_words

# Create an instance of CountVectorizer with Gujarati language support
vectorizer = CountVectorizer(ngram_range=ngram_range, analyzer='word', tokenizer=custom_tokenizer, token_pattern=r'\b\w+\b')

# Fit the vectorizer to the filtered sentences and transform them into n-grams
ngrams = vectorizer.fit_transform(filtered_sentences)

# Get the vocabulary (feature names) from the vectorizer
vocabulary = vectorizer.vocabulary_

# Create a dictionary to store the n-grams and their frequencies
ngram_dict = {}

# Iterate over the vocabulary and their corresponding counts
for feature in vocabulary:
    count = ngrams[:, vectorizer.vocabulary_.get(feature)].sum()
    ngram_dict[feature] = count
# Create a DataFrame from the n-gram dictionary
df_ngrams = pd.DataFrame(list(ngram_dict.items()), columns=['Words', 'Count'])

# Remove existing brackets '[' and ']' from the 'Words' column
df_ngrams['Words'] = df_ngrams['Words'].str.replace(r'\[|\]', '')

# Add brackets '[ ]' to every row in the 'Words' column
df_ngrams['Words'] = '[' + df_ngrams['Words'] + ']'

# Remove the comma from the second word or before the closing bracket ']'
df_ngrams['Words'] = df_ngrams['Words'].str.replace(r',(\w+)?\]', ']')

# Remove rows containing stopwords from the DataFrame
df_ngrams = df_ngrams[~df_ngrams['Words'].apply(lambda x: any(word in x for word in stop_words))]

# Save the DataFrame to a CSV file
df_ngrams.to_csv('ngrams_counts.csv', index=False)
```

```python
def read_words_from_file(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        text = file.read()
        words = text.split()
    return words


filename = '/content/drive/MyDrive/sentiment_accuracy/ngram_dict.txt'  # replace with your actual file name
word_list = read_words_from_file(filename)

import csv


def read_csv_to_list(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        reader = csv.reader(file)
        data = list(reader)
    return data


filename2 = '/content/ngrams_counts.csv'  # replace with your actual file name
data_list = read_csv_to_list(filename2)
data_list = data_list[1:]

import ast


# Function to convert string representation of lists into actual lists
def convert(item):
    return ast.literal_eval(item) if isinstance(item, str) and item.startswith('[') else item


# Use list comprehension to convert all string representations of lists into actual lists
converted_list = [[convert(item) for item in sublist] for sublist in data_list]

output = []
for i in converted_list:
    if i[0][0] in word_list:
        if i[0][0] == 'વધુ' and i[0][1] == 'માહિતી':
            continue
        data = [[i[0][0]+":"+i[0][1], i[1]]]
        output += data

print(output)
with open('output.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['Words', 'Count'])  # write header
    for row in output:
        writer.writerow(row)
```

```python
import csv

input_file = '/content/your_file_modified.csv'

# Open the input CSV file for reading and writing
with open(input_file, 'r+', newline='') as csv_file:
    reader = csv.DictReader(csv_file)
    headers = reader.fieldnames

    # Create a list to store the modified rows
    modified_rows = []

    for row in reader:
        words = row['Words'].split(',')
        modified_words = ["'{}'".format(word.strip()) for word in words]
        modified_row = row.copy()
        modified_row['Words'] = ','.join(modified_words)
        modified_rows.append(modified_row)

    # Move the file pointer to the beginning of the file
    csv_file.seek(0)

    # Write the modified rows back to the same CSV file
    writer = csv.DictWriter(csv_file, fieldnames=headers)
    writer.writeheader()
    writer.writerows(modified_rows)

    # Truncate any remaining data in the file
    csv_file.truncate()

print("CSV file has been modified.")
```
